```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
from tensorflow.keras.optimizers import SGD, Adam
import matplotlib.pyplot as plt
```

```python
x_data=np.array([[1,2,0],[5,4,3],[1,2,-1],[3,1,0],[2,4,2],[4,1,2],[-1,3,2],[4,3,3],[0,2,6],[2,2,1],[1,-2,-2],[0,1,3],[1,1,3],[0,1,4],[2,3,3]])
y_data=np.array([-3,5,-5,4,-3,10,-6,6,7,1,5,4,6,6,2])
```

```python
x_data.shape, y_data.shape
```

```python
model=Sequential()
```

```python
#model.add(Flatten(input_shape=(3,)))
#model.add(Dense(1,activation='linear'))
model.add(Dense(1,input_shape=(3,),activation='linear'))
```

```python
model.compile(optimzer=SGD(learning_rate=0.03),loss='mse')
```

```python
model.summary()
```

```python
hist=model.fit(x_data,y_data,epochs=1000)
```

```python
test_data=[[1,2,3],[5,4,6],[3,4,1]]
```

```python
pred=model.predict(np.array(test_data))
```

```python
real_val=[2*data[0]-3*data[1]+2*data[2]+1 for data in test_data]
```

```python
pred, real_val
```

```python
result=model.evaluate(test_data,real_val)
```

```python
result
```

```python
print(model.weights)
```

```python
plt.title('Loss Trend')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid
plt.plot(hist.history['loss'],label='train loss')
plt.legend(loc='best')
plt.show()
```
